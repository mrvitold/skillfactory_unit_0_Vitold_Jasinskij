```python
# starting position for the game
matrix = [
   ['-', '-', '-'],
   ['-', '-', '-'],
   ['-', '-', '-']
]

# range for filling coordinates
valid_range = range(3) 

# prints game matrix
def matrix_print():
    for i in matrix:
        print(i)

# checks if X wins by column. Counts X in each column        
def X_win_by_column():
    first_column = []
    first_column.count('X')
    for a in matrix:
        first_column.append(a[0]) 
    
    second_column = []
    second_column.count('X')
    for a in matrix:
        second_column.append(a[1])
    
    third_column = []
    third_column.count('X')
    for a in matrix:
        third_column.append(a[2])
    
# if there are 3 'X' in one of the column - X wins
    X_in_columns = [first_column.count('X'), second_column.count('X'), third_column.count('X')]
    if 3 in X_in_columns:
        return True

# checks if X wins by row. If there are 3 'X' in one of the row - X wins
def X_win_by_row():
    first_row_X_count = matrix[0].count('X')
    second_row_X_count = matrix[1].count('X')
    third_row_X_count = matrix[2].count('X')
    if 3 in [first_row_X_count, second_row_X_count, third_row_X_count]:
        return True

# checks if X wins by diagonal
def X_win_by_diagonal():
    if (matrix[0][0] == 'X' and  matrix[1][1] == 'X' and matrix[2][2] == 'X') or (matrix[0][2] == 'X' and  matrix[1][1] == 'X' and matrix[2][0] == 'X'):
        return True    
    
    
# checks if Zero wins by column
def Zero_win_by_column():
    first_column = []
    first_column.count('Zero')
    for a in matrix:
        first_column.append(a[0]) 
    
    second_column = []
    second_column.count('0')
    for a in matrix:
        second_column.append(a[1])
    
    third_column = []
    third_column.count('0')
    for a in matrix:
        third_column.append(a[2])
    
    Zero_in_columns = [first_column.count('0'), second_column.count('0'), third_column.count('0')]
    if 3 in Zero_in_columns:
        return True

# checks if Zero wins by row
def Zero_win_by_row():
    first_row_Zero_count = matrix[0].count('0')
    second_row_Zero_count = matrix[1].count('0')
    third_row_Zero_count = matrix[2].count('0')
    if 3 in [first_row_Zero_count, second_row_Zero_count, third_row_Zero_count]:
        return True

# checks if Zero wins by diagonal
def Zero_win_by_diagonal():
    if (matrix[0][0] == '0' and  matrix[1][1] == '0' and matrix[2][2] == '0') or (matrix[0][2] == '0' and  matrix[1][1] == '0' and matrix[2][0] == '0'):
        return True

# checks if X or Zero wins.
def win_solution():
    if X_win_by_column() or X_win_by_row() or X_win_by_diagonal():
        print('Xies won!')
        return True
    elif Zero_win_by_column() or Zero_win_by_row() or Zero_win_by_diagonal():
        print('Zeroes won!')
        return True
    

def game():
    while not win_solution(): # asks to input coordinates till X or Zero wins
        X_row = int(input('enter X row coordinate:'))
        X_column = int(input('enter X column coordinate:'))
        while (X_row and X_column) not in valid_range or matrix[X_row][X_column] != '-': # checks if entered coordinates are in range 0-2 or not already filled
            print('bad coordinates, choose another')
            X_row = int(input('enter X row coordinate:')) 
            X_column = int(input('enter X column coordinate:'))
        matrix[X_row][X_column] = 'X' # fills entered X into matrix
        matrix_print()
        if not win_solution(): # asks to input Zero coordinates if X has not won already 
            Zero_row = int(input('enter Zero row coordinate:'))
            Zero_column = int(input('enter Zero column coordinate:'))
            while (Zero_row and Zero_column) not in valid_range or matrix[Zero_row][Zero_column] != '-':
                print('bad coordinates, choose another')
                Zero_row = int(input('enter Zero row coordinate:')) 
                Zero_column = int(input('enter Zero column coordinate:'))
            matrix[Zero_row][Zero_column] = '0'
            matrix_print()      
        
game()
```
